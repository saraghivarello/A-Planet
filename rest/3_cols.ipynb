```python
def vec(c,mode):
    col='If you were to make this trip using a different mode of travel than your usual mode, which would be the best alternative?'
    v=c[c[col]==mode].groupby(['How would you rate your alternative?']).count()['mod']/len(c[c[col]==mode])*100
    v_=v.append(pd.Series({'Almost as good as my current choice':0,'Somewhat worse than my current choice':0,'Much worse than my current choice':0,'I do not know':0}))
    v_=v_.loc[['Almost as good as my current choice','Somewhat worse than my current choice','Much worse than my current choice','I do not know']]
    v_=v_[~v_.index.duplicated()]
    return v_
```
